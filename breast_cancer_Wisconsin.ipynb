```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
X_train.shape
```

```python
new_model = Sequential()



new_model.add(Dense(units=30,activation='relu'))

new_model.add(Dense(units=15,activation='relu'))


new_model.add(Dense(units=1,activation='sigmoid'))

# binary classification problem
new_model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
new_model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1
          )
```

```python
result=pd.DataFrame(data=new_model.history.history)
```

```python
result.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
new_model = Sequential()
new_model.add(Dense(units=30,activation='relu'))
new_model.add(Dense(units=15,activation='relu'))
new_model.add(Dense(units=1,activation='sigmoid'))
new_model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
new_model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model_loss = pd.DataFrame(new_model.history.history)
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
new_model = Sequential()
new_model.add(Dense(units=30,activation='relu'))
new_model.add(Dropout(0.5))

new_model.add(Dense(units=15,activation='relu'))
new_model.add(Dropout(0.5))

new_model.add(Dense(units=1,activation='sigmoid'))
new_model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
new_model.fit(x=X_train,y=y_train,
          epochs=600,validation_data=(X_test,y_test),
          verbose=1,
          callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(new_model.history.history)
model_loss.plot()
```

```python

predictions =new_model.predict(X_test)
predicted_classes = (predictions > 0.5).astype(int)
```

```python
predictions
```

```python
print(predicted_classes)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predicted_classes))
```
